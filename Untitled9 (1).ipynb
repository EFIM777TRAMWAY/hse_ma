```python
# Прогонка по пс2
bb=list(dataset.columns.values)
bb
y = dataset.iloc[:, 2].values
a = len(y)
dci = 0 # Цикл названия консолей
i = 0 # Рабочий цикл
tv = 0 #Количество проданных копий в млн
year = 0 #Год релиза
fl = []
demanded_consoles = np.array(['PS','PS2','PS3','PSP','GBA','DS','Wii','3DS','XB','X360', 'N64', 'GC'])
dc_len = len(demanded_consoles)
dc_len
while dci < dc_len:
    curcon = demanded_consoles[dci]
    while i < a: 
        v = dataset.values[i][2]
        if v == curcon:
            m = dataset.values[i][10]
            print(m)
            if m >= tv:    
                tv = m
                year = dataset.values[i][3]
                di = 0
                dlen = len(df)
                
                while di < dlen:
                    vdata = df.values[di][0]
                    if vdata == curcon:
                        df.set_value(di, 'MV Title Release', year)
                        df.set_value(di, 'Sold', tv)
                        df.set_value(di, 'Age_of_console', year - df.values[di][1])
                        df.set_value(di, 'Title', dataset.values[i][1])
                        i = a
                    di = di + 1
                    
        i = i+1
    dci = dci + 1
    i = 0
    tv = 0
```

Зададим дополнительный вопрос - а можно ли предсказать конец жизненного цикла консоли, зная дату релиза (предположительно) самой продаваемой игры на ней?

```python
df = df.assign(Life_Cycle=df['Discontinued'])
i = 0
a = len(df)
while i < a:
    df.set_value(i, 'Life_Cycle', df.values[i][6] - df.values[i][1])
    i = i + 1
df
```

```python
from sklearn.externals.six import StringIO
from sklearn import tree
from IPython.display import Image
import pydotplus as pydot
targetColumn = ['Life_Cycle']
FeatureColumns_dt2 = ['Age_of_console']
y = df[targetColumn].values
X = df[FeatureColumns_dt2].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 1)
clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
clf_tree.fit(X_train, y_train)
predicted = clf_tree.score(X_test, y_test)
print(predicted)
```

```python
dot_data = StringIO()
tree.export_graphviz(clf_tree, out_file=dot_data)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
